```python
for i in ['Vaksinasi 2_LansiaDKI']:  
    Q1  = df[i].quantile(0.25)
    Q3  = df[i].quantile(0.75)
    IQR = Q3-Q1

    Lower_Bound = Q1-(1.5*IQR)
    Upper_Bound = Q3+(1.5*IQR)

print('Kolom',i,'Memiliki Outliers sebesar:',len(df[(df[i]>Upper_Bound)|(df[i]<Lower_Bound)])/len(df)*100,'%')
Outliers_Vaksinasi_Lansia_DKI = df[(df[i]>Upper_Bound)|(df[i]<Lower_Bound)]
print(Outliers_Vaksinasi_Lansia_DKI['Vaksinasi 2_LansiaDKI'].mean())
print(df['Vaksinasi 2_LansiaDKI'].mean())
```

```python
for i in ['Positif Harian (Jakarta)']:  
    Q1  = df[i].quantile(0.25)
    Q3  = df[i].quantile(0.75)
    IQR = Q3-Q1

    Lower_Bound = Q1-(1.5*IQR)
    Upper_Bound = Q3+(1.5*IQR)

print('Kolom',i,'Memiliki Outliers sebesar:',len(df[(df[i]>Upper_Bound)|(df[i]<Lower_Bound)])/len(df)*100,'%')
Outliers_Positif_Harian_Jkt = df[(df[i]>Upper_Bound)|(df[i]<Lower_Bound)]
print(Outliers_Positif_Harian_Jkt['Positif Harian (Jakarta)'].mean())
print(df['Positif Harian (Jakarta)'].mean())
```
